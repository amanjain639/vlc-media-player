```python
import os
import pathlib

import time
import platform
from tkinter import ttk
from tkinter.filedialog import askopenfilename
from tkinter import *
from tkinter import font as tkFont
from tkinter import filedialog
from PIL import Image
from PIL import ImageTk
from PIL import ImageOps
import glob
import os,re
import numpy as np
import math
from pathlib import Path
from pynput.keyboard import Listener,Key
from win32api import GetKeyState
import time
import win32api
import win32con
import vlc
from threading import Timer,Thread,Event
file=None
pic=[".bmp",".pbm",".pgm",".ppm",".sr",".ras",".jpeg",".jpg",".jpe",".tiff",".tif",".png",".jfif"]
vid=[".mp4",".mp3",".mkv",".avi",".3gp",".3ga",".amv",".dts",".flv",".m4p",".m1v",".m2v",".mp1",".mp2",".mp3",
    ".mpeg1",".mpeg2",".mpeg4",".mpega",".rec",".tta",".vob",".voc",".wmv"]
#def main(filename):
#    file=filename
root = Tk()
allpics=[]
ind=0
sizee=0
file=""
###############################
class ttkTimer(Thread):
    def __init__(self, callback, tick):
        Thread.__init__(self)
        self.callback = callback
        self.stopFlag = Event()
        self.tick = tick
        self.iters = 0

    def run(self):
        while not self.stopFlag.wait(self.tick):
            self.iters += 1
            self.callback()
            #print("ttkTimer start")

    def stop(self):
        self.stopFlag.set()

    def get(self):
        return self.iters

class Player(Frame):
    def __init__(self, parent, title=None):
        Frame.__init__(self, parent)
    
        self.parent = parent
        if title == None:
            title = "tk_vlc"
        self.parent.title(title)
        self.player = None
        self.videopanel = ttk.Frame(self.parent)
        self.canvas = Canvas(self.videopanel).pack(fill=BOTH,expand=1)
        self.videopanel.pack(fill=BOTH,expand=1)
        

        ctrlpanel = ttk.Frame(self.parent)
        pause  = ttk.Button(ctrlpanel, text="Pause", command=self.OnPause)
        play   = ttk.Button(ctrlpanel, text="Play", command=self.OnPlay)
        stop   = ttk.Button(ctrlpanel, text="Stop", command=self.OnStop)
        self.mute = StringVar()
        self.mute.set("Mute")
        volume = ttk.Button(ctrlpanel, textvariable=self.mute, command=self.OnToggleVolume)
        openn = ttk.Button(ctrlpanel, text="Open", command=self.OnOpen)
        leftt = ttk.Button(ctrlpanel, text="<", command=self.OnLeft)
        rightt = ttk.Button(ctrlpanel, text=">", command=self.OnRight)
        leftt.pack(side=LEFT)
        openn.pack(side=LEFT)
        pause.pack(side=LEFT)
        play.pack(side=LEFT)
        stop.pack(side=LEFT)
        rightt.pack(side=LEFT)
        volume.pack(side=LEFT)
        self.volume_var = IntVar()
        self.volslider = Scale(ctrlpanel, variable=self.volume_var, command=self.volume_sel, 
                from_=0, to=100, orient=HORIZONTAL, length=100)
        self.volslider.pack(side=LEFT)
        self.volume_var.set(100) 
        ctrlpanel.pack(side=BOTTOM)
        ctrlpanel2 = ttk.Frame(self.parent)
        self.scale_var = DoubleVar()
        self.timeslider_last_val = ""
        self.timeslider = Scale(ctrlpanel2, variable=self.scale_var, command=self.scale_sel, 
                from_=0, to=1000, orient=HORIZONTAL, length=1000)
        self.timeslider.pack(side=BOTTOM, fill=X,expand=1)
        self.timeslider_last_update = time.time()
        ctrlpanel2.pack(side=BOTTOM,fill=X)
        self.Instance = vlc.Instance()
        self.player = self.Instance.media_player_new()
        self.timer = ttkTimer(self.OnTimer, 1.0)
        self.timer.start()
        self.parent.update()


    def OnExit(self, evt):

        self.Close()

    def OnOpen(self):
        self.OnStop()
        global file
        global ind
        select=Tk()
        
        
        file=filedialog.askopenfilename()
        if (file.lower().endswith(tuple(pic)) or file.lower().endswith(tuple(vid))):
            self.Media = self.Instance.media_new(file)
            self.player.set_media(self.Media)
            self.player.set_hwnd(self.GetHandle())
            select.destroy()
            allpics=[]
            findd()
            self.OnStop()
            if self.player.play() == -1:
                self.errorDialog("Unable to play.")

    def OnPlay(self):

        if not self.player.get_media():
            self.OnOpen()
        else:
            if self.player.play() == -1:
                self.errorDialog("Unable to play.")

    def GetHandle(self):
        return self.videopanel.winfo_id()

    def OnPause(self):
        self.player.pause()

    def OnStop(self):
        self.player.stop()
        self.timeslider.set(0)

    def OnTimer(self):
        if self.player == None:
            return
        length = self.player.get_length()
        dbl = length * 0.001
        self.timeslider.config(to=dbl)

        tyme = self.player.get_time()
        if tyme == -1:
            tyme = 0
        dbl = tyme * 0.001
        self.timeslider_last_val = ("%.0f" % dbl) + ".0"
        #print(self.scale_var.get())
        if time.time() > (self.timeslider_last_update + 0.001):
            self.timeslider.set(dbl)

    def scale_sel(self, evt):
        if self.player == None:
            return
        nval = self.scale_var.get()
        sval = str(nval)
        if self.timeslider_last_val != sval:
            self.timeslider_last_update = time.time()
            mval = "%.0f" % (nval * 1000)
            self.player.set_time(int(mval))


    def volume_sel(self, evt):
        if self.player == None:
            return
        volume = self.volume_var.get()
        if volume > 100:
            volume = 100
        if self.player.audio_set_volume(volume) == -1:
            self.errorDialog("Failed to set volume")



    def OnToggleVolume(self):
        is_mute = self.player.audio_get_mute()
        if(is_mute==True):
            self.mute.set("Mute")
        else:
            self.mute.set("Umute")
        self.player.audio_set_mute(not is_mute)
        self.volume_var.set(self.player.audio_get_volume())

    def OnSetVolume(self):
        volume = self.volume_var.get()
        print("volume= ", volume)
        if volume > 100:
            volume = 100
        if self.player.audio_set_volume(volume) == -1:
            self.errorDialog("Failed to set volume")

    def errorDialog(self, errormessage):
        edialog = tkMessageBox.showerror(self, 'Error', errormessage)
        ##############################################

    
    def OOpenn(self):
        global ind
        file=allpics[ind]
        if (file.lower().endswith(tuple(pic)) or file.lower().endswith(tuple(vid))):
            self.Media = self.Instance.media_new(file)
            self.player.set_media(self.Media)
            self.player.set_hwnd(self.GetHandle())
            self.OnStop()
            if self.player.play() == -1:
                self.errorDialog("Unable to play.")
            self.OnTimer()
            
    
    def OnLeft(self):
        global ind
        if(ind>0):
            ind=ind-1
            file=allpics[ind]
        else:
            ind=sizee-1
            file=allpics[ind]
        self.OOpenn()

    def OnRight(self):
        global ind
        if(ind==sizee-1):
            ind=0
            file=allpics[ind]
        else:
            ind=ind+1
            file=allpics[ind]
        mval = "%.0f" % (1 * 1000)
        dbl=0.001
        self.player.set_time(int(mval))
        self.timeslider_last_val = ("%.0f" % dbl) + ".0"
        self.OOpenn()
def Tk_get_root():
    if not hasattr(Tk_get_root, "root"): #(1)
        Tk_get_root.root= Tk()
    return Tk_get_root.root
        
def _quit():
    print("_quit: bye")
    root = Tk_get_root()
    root.quit()     
    root.destroy()
    os._exit(1)

###############################
root.configure(bg='black')
panelPic=None
def findd():
    file_pattern = re.compile(r'.*?(\d+).*?')
    def get_order(file):
        match = file_pattern.match(Path(file).name)
        if not match:
            return math.inf
        return int(match.groups()[0])

    
    global ind
    global file
    global allpics
    global sizee
    folder=file.split('/')
    folder=folder[:-1]
    folder='/'.join(folder)
    folder=folder+'/'
    files= glob.glob(folder+'*')
    allpics=[]
    j=0
    sizee=0
    for i in files:
        i=i.replace('\\','/')
        if(i.lower().endswith(tuple(pic))or i.lower().endswith(tuple(vid))):
            allpics.append(i)
            j=j+1
    #print(allpics)
    sizee=j
    sorted_files = sorted(allpics, key=get_order)
    ind=allpics.index(file)
    #print(allpics,ind)

    
def showPic():
    
   
    
    global ind
    global file
    global allpics
    ShowPic = Tk_get_root()
    ShowPic.protocol("WM_DELETE_WINDOW", _quit)
    ShowPic.configure(bg='black')
    ShowPic.geometry("1920x1080")
    ShowPic.attributes('-fullscreen',True)
    player = Player(ShowPic, title="tkinter vlc")
    ShowPic.update()
    ShowPic.mainloop()
    
def Select():
    global file
    global ind
    select=Tk()
    allpics=[]
    file=filedialog.askopenfilename()
    if (file.lower().endswith(tuple(pic)) or file.lower().endswith(tuple(vid))):
        select.destroy()
        findd()
        showPic()
root.geometry("1920x1080")
def Open():
    root.destroy()
    showPic()
root.attributes('-fullscreen',True)
fontt = tkFont.Font(family='Comic Sans MS', size=60, weight='bold')
exit_button = Button(root, text="Open",font=fontt,
                     command=Open,height = 100, width = 30,activeforeground = "red",activebackground = "blue",bg="red"
                     ,fg="blue") 
exit_button.pack(pady=100)
root.update()
root.deiconify()
root.mainloop()
```
